```sql
/****** Object:  StoredProcedure [dbo].[sotrudnik_otchet]    Script Date: 29.11.2021 20:48:14 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sotrudnik_otchet]
as
begin
	declare @Fname varchar(50),@id_sotrud int,@Nazva_klient varchar(50),@id_klient int,@Nazva_t varchar(50),@SumPrice int,@Zakazid int
	declare sotr cursor for Select [Fname],[id_sotrudnik] from dbo.Sotrudnik
	open sotr
	fetch next from sotr into @Fname,@id_sotrud
	while(@@FETCH_STATUS = 0)
	begin 
		print 'Сотрудник '+@Fname
		declare klient cursor for Select distinct id_klient ,Nazva from dbo.klient
		INNER JOIN dbo.Zakaz ON dbo.klient.id_klient = dbo.Zakaz.K_klien
		where K_sotrud = @id_sotrud 
		open klient
			fetch next from klient into @id_klient,@Nazva_klient
		while(@@FETCH_STATUS = 0)
		begin
			print '    Клиент ' + @Nazva_klient
			
			declare tovar cursor for Select [K_zakaz],[Nazva_t], SUM(kilk*price) from dbo.tovar
			LEFT JOIN dbo.zakaz_tovar ON dbo.tovar.id_tovar = dbo.zakaz_tovar.id_tovar 
			LEFT JOIN dbo.Zakaz ON dbo.zakaz_tovar.id_zakaz = dbo.Zakaz.K_zakaz
			where K_sotrud = @id_sotrud and K_klien = @id_klient
			group by K_zakaz,Nazva_t
			open tovar
			fetch next from tovar into @Zakazid,@Nazva_t,@SumPrice
			while(@@FETCH_STATUS = 0)
			begin
				print '        Nomer: '+Cast(@Zakazid as varchar(50))+', '+@Nazva_t+', Sum (' +Cast(@SumPrice as varchar(50))+')'
				fetch next from tovar into @Zakazid,@Nazva_t,@SumPrice
			end
			close tovar
			deallocate tovar
			fetch next from klient into @id_klient,@Nazva_klient
		end
	
		fetch next from sotr into @Fname,@id_sotrud
		close klient
		deallocate klient
	select @SumPrice = SUM(kilk * price) from dbo.Sotrudnik 
		INNER JOIN dbo.Zakaz ON dbo.Sotrudnik.id_sotrudnik = dbo.Zakaz.K_sotrud 
		INNER JOIN dbo.zakaz_tovar ON dbo.Zakaz.K_zakaz = dbo.zakaz_tovar.id_zakaz 
		INNER JOIN dbo.tovar ON dbo.zakaz_tovar.id_tovar = dbo.tovar.id_tovar
		where id_sotrudnik=@id_sotrud
		group by id_sotrudnik
		print ' Total Sum : '++cast(@SumPrice as varchar(50))

	end

	close sotr
	deallocate sotr
end;


exec sotrudnik_otchet
GO
```
